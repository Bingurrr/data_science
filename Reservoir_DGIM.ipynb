```python
# 비복원
import random

class Reservoir :
  def __init__(self, k) :
    self.sampled = []
    self.k = k
    self.cnt = 0
  
  def put(self, item) :
    if self.cnt < self.k :
      self.sampled.append(item)
    else :
      r = random.randint(0, self.cnt)
      if r < self.k :
        self.sampled[r] = item
    self.cnt += 1 

res = [0  for _ in range(1000)]
#print(reservoir.sampled)
for _ in range(10000) :
  reservoir = Reservoir(100)
  for i in range(1000) :
    reservoir.put(i)
    #print(reservoir.sampled)
  for j in reservoir.sampled :
    res[j] += 1
print(res)
```

```python
# 실습 결과물
class Bucket :
  def __init__(self, start, end) :
    self.start = start
    self.end = end

  def __repr__(self) :
    return f"({self.start},{self.end})"



class DGIM:
  def __init__(self) :
    self.bucket_tower = [[]]
    self.ts = 0

  def put(self, bit) :
    if bit == 1 :
      self.bucket_tower[0].insert(0, Bucket(self.ts, self.ts))

      layer = 0
      len(self.bucket_tower) < layer + 2
      while len(self.bucket_tower[layer]) > 2 :
        if len(self.bucket_tower) <= layer + 1 :
          self.bucket_tower.append([])

        b1 = self.bucket_tower[layer].pop()
        b2 = self.bucket_tower[layer].pop() 
        b1.end = b2.end

        self.bucket_tower[layer+1].insert(0,b1)
        layer += 1
      
    self.ts += 1
  
  def count(self, k) :
    s = self.ts - k
    cnt = 0 
    for layer, buckets in enumerate(self.bucket_tower) :
      for bucket in buckets :
        if s <= bucket.start :
          cnt += (1 << layer)
        elif s <= bucket.end :
          cnt += (1 << layer) * (bucket.end - s + 1) // (bucket.end - bucket.start + 1)
          return cnt
        else :
          return cnt
    return cnt      

dgim = DGIM()
bitstream = []
for i in range(10) :
  prob = random.random()
  for j in range(random.randint(20,50)) :
    if random.random() < prob :
      bitstream.append(1)
    else :
      bitstream.append(0)
for b in bitstream :
  dgim.put(b)
  print(dgim.bucket_tower)
for k in range(1, 200):
  print(k, dgim.count(k), sum(bitstream[-k:]))
```

```python
# 방법 1
"""
  0~15사이 수 입력이니까 4비트
  수업시간에 구현한 DGIM알고리즘을 자리수 별로 4번 DGIM을 수행하고
  각각 1,2,4,8을 곱해 합을 구함
"""
# 방법 2 
"""
  위에 구현한 DGIM2()를 사용해서 합을 구함
"""

n = random.randint(0,15)
n = format(n,'b')

# 합
sum_real = [] #실제 합
sum1 = [] #방법1로 구한 합
sum2 = [] #방법2로 구한 합

#실제 값
num = []

#방법 1
dgim1 = DGIM() # 1의자리
dgim2 = DGIM() # 10의 자리
dgim3 = DGIM() # 100의 자리
dgim4 = DGIM() # 1000의 자리

#방법2 
dgim_part = DGIM2()

k = 0
for _ in range(10000) :
  n = random.randint(0,15)
  #실제
  num.append(int(n))
  #방법2
  dgim_part.put(int(n))

  #방법 1을 수행하기 위해 정수를 4비트의 2진수로 변환
  #정수를 4비트로 변환
  n = format(n,'b')
  while len(n) != 4 :
    n = '0' + n
  #4비트니까 DGIM4번 돌린다
  #1의 자리
  dgim1.put(int(n[3]))
  #2의 자리
  dgim2.put(int(n[2]))
  #3의 자리
  dgim3.put(int(n[1]))
  #4의 자리
  dgim4.put(int(n[0]))

for k in range(1, 2001) :  
  #실제값
  sum_real.append(sum(num[-k:]))
  #방법1
  num_sum1 = dgim1.count(k) + 2 * dgim2.count(k) + 4 * dgim3.count(k) + 8 * dgim4.count(k)
  sum1.append(num_sum1)
  #방법2
  sum2.append(dgim_part.count(k))



print(sum_real)
print(sum1)
print(sum2) 
```
